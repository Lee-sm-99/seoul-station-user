```python
import pandas as pd

# 데이터 생성
city_names = pd.Series(['서울', '부산', '대전', '세종'])
population = pd.Series([9776000, 3429000, 1531000])

city_names1 = pd.Series(['서울', '일산', '대전', '세종'])
population1 = pd.Series([9776000, 3429000, 1531000,243532])

city_names2 = pd.Series(['서울', '부산', '대구', '세종'])
population2 = pd.Series([9776000, 3429000, 1531000,2345563])

city_names3 = pd.Series(['서울', '부산', '대전', '세종'])
population3 = pd.Series([9776000, 3429000, 1531000,234267])

# 데이터프레임 생성
df1 = pd.DataFrame({'City': city_names, 'Population': population})
df2 = pd.DataFrame({'City': city_names1, 'Population': population1})
df3 = pd.DataFrame({'City': city_names2, 'Population': population2})
df4 = pd.DataFrame({'City': city_names3, 'Population': population3})

# 데이터프레임을 리스트에 저장
data_frames = [df1, df2, df3, df4]

# 겹치는 도시와 인구 정보를 저장할 데이터프레임 초기화
result_df = pd.DataFrame(columns=['City', 'Population'])

# 첫 번째 데이터프레임을 기준으로 설정
result_df = df1

# 나머지 데이터프레임과 비교하며 겹치는 도시만 남김
for df in data_frames[1:]:
    result_df = result_df.merge(df, how='inner', on='City')

# 결과 출력
print(result_df)
```
